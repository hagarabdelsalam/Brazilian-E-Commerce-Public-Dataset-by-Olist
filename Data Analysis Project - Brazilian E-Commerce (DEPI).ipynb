##### <span style="color:orange">Use previous monthly revenue trends to predict future sales and revenue growth

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# 1. Convert 'order_delivered_customer_date' to datetime (if not already in datetime format)
df_clean['order_delivered_customer_date'] = pd.to_datetime(df_clean['order_delivered_customer_date'], errors='coerce')

# 2. Set 'order_delivered_customer_date' as the index for time series analysis
df_clean = df_clean.set_index('order_delivered_customer_date')

# 3. Resample the data monthly by summing the total_order_amount
monthly_revenue = df_clean['total_order_amount'].resample('M').sum()

# 4. Fit an ARIMA model to the monthly revenue data
# (The ARIMA order (1,1,1) is an example; you may need to fine-tune this order)
model = ARIMA(monthly_revenue, order=(1, 1, 1))
model_fit = model.fit()

# 5. Forecast future revenue for the next 12 months
forecast_steps = 12
forecast_result = model_fit.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=monthly_revenue.index[-1] + pd.offsets.MonthBegin(), 
                               periods=forecast_steps, freq='M')
forecast_series = pd.Series(forecast_result.predicted_mean, index=forecast_index)

# Plot historical monthly revenue and forecasted revenue
plt.figure(figsize=(12, 6))
plt.plot(monthly_revenue, label='Historical Monthly Revenue')
plt.plot(forecast_series, label='Forecasted Revenue', linestyle='--', color='red')
plt.title('Monthly Revenue Forecast')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()

print("Forecasted Monthly Revenue:")
print(forecast_series)

```

##### <span style="color:orange">Installment Payment Variation Analysis

```python
# Filter out rows where payment_installments is 0 to avoid division by zero
df_clean = df_clean[df_clean['payment_installments'] != 0]

# Calculate the installment amount using the formula:
# I = (price + freight_value) / payment_installments
df_clean['installment_amount'] = (df_clean['price'] + df_clean['freight_value']) / df_clean['payment_installments']

# Drop rows where installment_amount is NaN or infinite (if any)
installment_data = df_clean['installment_amount'].replace([float('inf'), -float('inf')], pd.NA).dropna()

# Variation Analysis: Calculate summary statistics for the installment amounts
average_installment = installment_data.mean()
std_installment = installment_data.std()
variance_installment = installment_data.var()

print("Average Installment Amount:", average_installment)
print("Standard Deviation of Installment Amount:", std_installment)
print("Variance of Installment Amount:", variance_installment)
```

##### <span style="color:orange"> Average Order Value by Seller

```python
# Create a new column for total order amount
df_clean['total_order_amount'] = df_clean['price'] + df_clean['freight_value']

# Group by seller_id, aggregate total revenue and order count, then compute Average Order Value
aov_by_seller = df_clean.groupby('seller_id').agg(
    total_revenue=('total_order_amount', 'sum'),
    order_count=('order_id', 'count')
)
aov_by_seller['average_order_value'] = aov_by_seller['total_revenue'] / aov_by_seller['order_count']

print("Average Order Value by Seller:")
print(aov_by_seller[['average_order_value']])
```

##### <span style="color:orange">Average Order Value (AOV) per Product

```python
revenue = df_clean.groupby('product_id').apply(
    lambda x: (x['price'] + x['freight_value']).sum())
order_count = df_clean.groupby('product_id')['order_id'].nunique()
aov_per_product = (revenue / order_count).sort_values(ascending=False)

print("Average Order Value per Product:")
print(aov_per_product)
```

##### <span style="color:orange">Return or Cancellation Rate per Product

```python
# Define not fulfilled orders based on order_status.
# Adjust the list if you have other statuses that indicate non-fulfillment.
df_clean['not_fulfilled'] = df_clean['order_status'].isin(['canceled', 'returned'])

# Calculate total number of orders per product
total_orders_per_product = df_clean.groupby('product_id')['order_id'].count()

# Calculate the number of not fulfilled orders per product
not_fulfilled_per_product = df_clean.groupby('product_id')['not_fulfilled'].sum()

# Calculate the return/cancellation rate per product as a percentage
cancellation_rate = (not_fulfilled_per_product / total_orders_per_product) * 100

# Sort the results in descending order (highest rate first)
cancellation_rate = cancellation_rate.sort_values(ascending=False)

print("Cancellation Rate per Product (in %):")
print(cancellation_rate)
```
